# Vibe coding 與除錯

「Vibe coding」是指在 LLM 協助下編程的暱稱。在 Cursor 中特別好用！但要做得好有些技巧，課程會涵蓋。以下是幾條要訣：

1. 若用 ChatGPT 或 Claude 等 LLM 寫程式，在 prompt 中加入當前日期，並要求使用最新版 API。對 OpenAI API 尤其重要！ChatGPT 常會用過時的自家 API 版本……

2. 在 prompt 中要求程式盡量簡短。LLM 常加一堆不必要的檢查，讓程式難讀難除錯。

3. 同一問題問多個 LLM，選最清楚簡潔的答案。

4. 類似做法：拿一個 LLM 的答案，請另一個 LLM 檢查正確性與是否可更簡潔。

5. 請 LLM 提供 3 種解法，選最清楚的一種。

## 較大專案的 Vibe Coding

盡量避免讓 LLM 一次產生 100–200 行以上程式；除非你已很熟悉，否則很難除錯。

改為：把問題拆成可獨立測試的小步驟。若不確定如何拆分——這也可以問 LLM！

對每個積木：
- 用上述技巧讓 LLM 寫程式
- 也請 LLM 寫測試驗證
- 自己測試，確認正確

這樣能更有信心地建構較大專案。
