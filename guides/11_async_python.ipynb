### 現在是一些範例：


````python
# 定義一個 async 函式

import asyncio

async def do_some_work():
    print("Starting work")
    await asyncio.sleep(1)
    print("Work complete")
````

````python
# 這會做什麼？

do_some_work()
````

````python
# 好，再試一次！

await do_some_work()
````

````python
# 這有什麼問題？

async def do_a_lot_of_work():
    do_some_work()
    do_some_work()
    do_some_work()

await do_a_lot_of_work()
````

````python
# 有趣的警告！來修復

async def do_a_lot_of_work():
    await do_some_work()
    await do_some_work()
    await do_some_work()

await do_a_lot_of_work()
````

````python
# 現在並行執行
# 重要的是這不是你可能習慣的「多執行緒」
# asyncio 在單一執行緒上執行，但用迴圈在一個任務等待時切換

async def do_a_lot_of_work_in_parallel():
    await asyncio.gather(do_some_work(), do_some_work(), do_some_work())

await do_a_lot_of_work_in_parallel()
````

### 最後——試寫一個 python 模組呼叫 do_a_lot_of_work_in_parallel

見頂部連結；模組中需要類似：

```python
if __name__ == "__main__":
    asyncio.run(do_a_lot_of_work_in_parallel())
```
