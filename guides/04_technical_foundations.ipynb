# 技術基礎

對我們使用的基本技術概念感到舒適至關重要。這會讓你整個課程體驗好得多——若不確定發生什麼會很挫折。

這些指南應建立對我們使用之底層技術的信心。


## 主題 1：ChatGPT 與 OpenAI API

### ChatGPT 與 OpenAI 提供的 GPT API 有何不同？

#### ChatGPT 是給終端使用者的工具。它是為 AI 使用者設計的聊天產品。
- 有免費方案，也有功能更多的付費訂閱。
- 訂閱方案讓使用者近乎無限使用聊天產品。

#### API 是給 AI 工程師——軟體工程師與資料科學家——建構其他商業產品的服務。
- 讓技術人員（你我）存取底層模型（如「GPT4.1」與「o3」）以建構自己的產品。
- 若願意，我們可用 API 建自己的 ChatGPT 版本，向終端使用者收費。
- 像多數 API，OpenAI 按 API 使用量收少量費用。課程多數用 gpt-4o-mini 的範例，約每次 API 呼叫 $0.001。

### 我為 ChatGPT 付 $20/月——為何 API 還要付？

- 希望現在清楚了。API 不是給消費者；是給工程師建可收費的平台。
- 若訂閱就能用 API，你就能以更便宜價格提供 ChatGPT 工具，讓 OpenAI 破產！
- 記住：每次 API 呼叫可能需 10,000,000,000,000 次浮點運算——那計算用電！

若不呼叫 API，可在本機跑開源模型，但通常計算量少 1,000 倍——即使很小，處理仍會計入電費。
