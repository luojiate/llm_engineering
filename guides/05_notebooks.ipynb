## 現在回去重新執行兩格前的 print 儲存格

看到即使 favorite_fruit 在 notebook 更下方被改變，它仍印出不同內容？

程式碼在 notebook 中的**出現順序**不重要。重要的是程式碼被**執行**的順序。此 notebook 背後有 python 程序在改變變數。

這讓一些初次用 notebook 的人困惑。


````python
# 然後執行此儲存格兩次，看你是否理解發生什麼

print(f"My favorite fruit is {favorite_fruit}")

favorite_fruit = "apples"
````

# 解釋「核心（Kernel）」

此 notebook 背後有 Python 程序，在你執行時執行每個儲存格。該 Python 程序稱為核心（Kernel）。每個 notebook 有各自獨立的核心。

你可點上方「Restart Kernel」。

若然後試執行下一儲存格，會出錯，因為 favorite_fruit 不再定義。需從 notebook 頂部重新執行儲存格。然後下一儲存格應可正常執行。


````python
print(f"My favorite fruit is {favorite_fruit}")
````

# 使用 markdown

用 Markdown 格式加一些註解，或從這裡複製貼上：

```
# This is a heading
## This is a sub-head
### And a sub-sub-head

I like Jupyter Lab because it's
- Easy
- Flexible
- Satisfying
```

在儲存格按 Shift+Return 即可變成格式化文字。
點儲存格按垃圾桶圖示可移除。

# 課程中會遇到的小東西

這不一定是 notebook 功能，但是值得知道的套件，在 notebook 中很有用。

套件 `tqdm` 若包住任何可迭代物件會印出漂亮進度條。


````python
# 這是沒有進度條的程式碼
# 會花 10 秒讓你納悶發生什麼……

import time

spams = ["spam"] * 1000

for spam in spams:
    time.sleep(0.01)
````

````python
# 現在，有漂亮小進度條：

import time
from tqdm import tqdm

spams = ["spam"] * 1000

for spam in tqdm(spams):
    time.sleep(0.01)
````

````python
# 換個主題，這是用 markdown 印輸出的實用方式

from IPython.display import Markdown, display

display(Markdown("# This is a big heading!\n\n- And this is a bullet-point\n- So is this\n- Me, too!"))
````
