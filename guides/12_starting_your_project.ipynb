## 如何開始實作你的想法

你有了想打造的點子。

恭喜！最重要的部分已經完成。點子就是一切！

但現在不確定從何開始。我常被問到這個。

我想給你三條關鍵建議。


## 建議 1：做個科學家

AI 工程師要戴兩頂帽子：軟體工程師與資料科學家。

專案初期，務必摘下軟體工程師的帽子，戴上資料科學家的帽子。

許多軟體工程背景的人（包括我！）會掙扎於此。常見是回到舒適區，問架構、設計、資料管線、擴展性、部署。

這些都是好問題，但不是早期階段的關鍵。它們不是難題，不會決定點子成敗。

關鍵問題需要科學家思維。

問自己：  
1. 如何評估模型成功？量測什麼？  
2. 你有哪些資料、還需要什麼？  
3. 如何建小原型測試不同技術、了解模型表現？

先處理科學問題——用 LLM 能達成什麼、如何達成。

建議先在 Notebook 工作，再進 python 模組，以強迫自己保持科學家心態。

Agent 專案更是如此。容易畫大架構圖說明 agents 如何協作，那是軟體工程師思維。建議換方式：實驗不同做法、調查什麼有效、驗證假設並迭代。

## 建議 2：擁抱研發（R&D）

與建議 1 類似，但值得重複！

常被問：「Ed，我想做某方案，該用模型 A、B 還是 C？該用 Agents、RAG 還是 fine-tuning？」

我幾乎總是回答：  
- 全都試！關鍵是實驗；(a) 訂評估標準，(b) 建立精選資料集，(c) 測試不同想法並比較表現
- 我常有直覺哪個最好，但常錯！別信我：自己試  
- 實驗無可取代。

## 建議 3：夢想要大，但從**小**做起！

也常有人寄 200–300 行程式說「Ed，這表現不好，請修」。😂

如除錯指南所說——不是這樣運作的！

專案初期務必從小從簡開始。打磨每個 prompt；細究每一步。確認每次 LLM 呼叫如你所願，迭代輸入直到回應穩定可靠。

若始終以小而可測的積木增量工作，你應能完全掌握狀況。

當然我樂意協助專案、修問題。但若增量進行，你總能清楚哪裡表現不佳——那樣我幫你會非常高效。

# 總結

你可能注意到三條建議的共同主題。非 Data Science 背景者可能覺得這種 ad-hoc 方式不適、感覺「草莽」不滿足。但我認為，成為成功 AI 工程師最重要的技能是：對不確定感到自在、享受實驗、擁抱科學家身份。
