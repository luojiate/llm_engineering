### 除錯公式：用 10 行程式重現

優秀除錯有方法也有藝術，多練習就會！要點如下：

1. 重現  
穩定重現你的問題

2. 簡化  
把問題縮到最簡的重現方式。太多人寄 100+ 行程式說「Ed，這不行，請修」。不是這樣運作的！我會先縮到核心——理想上 10–20 行以內。我曾示範如何用 2 行重現！

_重點是，這通常比別人幫你做更容易。_ 雖然枯燥，但幾乎總是最佳起點。縮到幾行程式後，你常會自己發現問題！

3. 求助  
用幾行程式重現後，若仍看不懂——該求助了！可先問 ChatGPT 等；給他們短範例與完整 stack trace。

也可聯絡我！我會盡力讓你的學習體驗更好；若卡住，我會幫你解困。
