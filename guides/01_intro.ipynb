# 歡迎來到 Guides！

## 簡介

本課程設計給多元背景學員。若你對部分主題較陌生，我準備了技術簡報幫你建立基礎。

這些材料供自學使用；你可依自己的步調閱讀、實驗。

我大量借助 AI 協助撰寫指南，並盡量讓內容實用且與課程相關。

唯一要求是：若某主題對你完全陌生，請先閱讀對應指南再進入該週內容。這會大幅提升學習體驗。

若你已是資深工程師，可略過已知主題；但即使如此，快速瀏覽仍可能有收穫。
